## 1.2 Creaci√≥n de nuevas Caracteristicas [2 Puntos] 

Como ya se les comento, Mr. Lepin esta interesado en obtener las caracter√≠sticas **LRMFP**, para esto les se√±ala que estas caracter√≠sticas se construyen en base a las siguientes definiciones:

- **Length (L)**: Intervalo de tiempo, en d√≠as, entre la primera y la √∫ltima visita del cliente. Mientras mas grande sea el valor, mas fiel es el cliente.
- **Recency (R)**: Indica la actualidad de la interacci√≥n de un cliente con la empresa, y da informaci√≥n sobre la tendencia a repetir la compra. Se define como: 
$$Recency(n)=\dfrac{1}{n} \sum^n_{i=1} date\_diff(t_{fecha final}, t_{m-i+1})$$

 Donde $date\_diff$ representa la diferencia en d√≠as entre la fecha de finalizaci√≥n del periodo de observaci√≥n ($t_{fecha final}$), y la fecha de una visita del cliente cercana a $t_{fecha final}$, $t_{m-i+1}; t_{m}$ es la √∫ltima visita del cliente; y n es el n√∫mero de visitas recientes del cliente consideradas.

- **Monetary (M)**: El t√©rmino "monetario" se refiere a la cantidad media de dinero gastada por cada visita del cliente durante el per√≠odo de observaci√≥n y refleja la contribuci√≥n del cliente a los ingresos de la empresa.

- **Frequency (F)**: Se refiere al n√∫mero total de visitas del cliente durante el periodo de observaci√≥n. Cuanto mayor sea la frecuencia, mayor ser√° la fidelidad del cliente. 

- **Periodicity (P)**: Representa si los clientes visitan las tiendas con regularidad.

$$Periodicity(n)=std(IVT_1, ..., IVT_n)$$

&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Donde $IVT$ denota el tiempo entre visitas y n representa el n√∫mero de valores de tiempo entre visitas de un cliente.
 

$$IVT_i=date\_diff(t_{i+1},t)$$

En base a las definiciones se√±aladas, dise√±e una funci√≥n que permita obtener las caracter√≠sticas **LRMFP** recibiendo un DataFrame como entrada. Para esto, no estar√° permitido el uso de iteradores, utilice todas las herramientas que les ofrece `pandas` para realizar esto.

Una referencia que le puede ser √∫til es el [documento original](https://www.researchgate.net/publication/315979555_LRFMP_model_for_customer_segmentation_in_the_grocery_retail_industry_a_case_study) en donde se propone este m√©todo.

**Nota:** Para la $fechafinal$ utilice la fecha m√°xima del dataset m√°s 1 d√≠a.

**Ejemplo de Resultado Esperado:**

| Customer ID | Length | Recency | Frequency | Monetary | Periodicity |
|------------:|-------:|--------:|----------:|---------:|------------:|
|   12346.0   |    294 |      67 |        46 |   -64.68 |        37.0 |
|   12347.0   |     37 |       3 |        71 |  1323.32 |         0.0 |
|   12349.0   |    327 |      43 |       107 |  2646.99 |        78.0 |
|   12352.0   |     16 |      11 |        18 |   343.80 |         0.0 |
|   12356.0   |     44 |      16 |        84 |  3562.25 |        12.0 |

**Respuesta:**

```python
def custom_features(dataframe_in):
    df = dataframe_in.copy()
    df['Money'] = df['Quantity'] * df['Price']
    df = df.drop(columns=['Invoice', 'StockCode', 'Description', 'Quantity', 'Price', 'Country'])
    df['Customer ID'] = df['Customer ID'].astype('int')
    df = df.groupby(['InvoiceDate', 'Customer ID'], as_index=False).sum()
    fecha_final = df['InvoiceDate'].max() + np.timedelta64(1, 'D')
    df = df[df.duplicated(subset=['Customer ID'], keep=False)]
    output = pd.DataFrame()
    gb = df.groupby('Customer ID', as_index=False)
    for name, group in gb:
        l = (group['InvoiceDate'].max() - group['InvoiceDate'].min()).days
        r = (fecha_final - group['InvoiceDate'].iloc[-1:]).mean().days
        f = len(group)
        m = round(group['Money'].mean(), 2)
        b = group.iloc[1:].copy().reset_index(drop=True)
        a = group.iloc[:-1].copy().reset_index(drop=True)
        p = (b['InvoiceDate'] - a['InvoiceDate']).std().days
        row = pd.DataFrame([[name, l, r, f, m, p]], columns=['Customer ID', 'Length', 'Recency', 'Frequency', 'Monetary', 'Periodicity'])
        output = pd.concat([output, row], ignore_index=True)
    return output.fillna(0)
```

```python
output = custom_features(df_retail)
output
```

**Observaciones:** Obtuve resultados ligeramente distintos a los del ejemplo. La diferencia de resultados en "Length" no me lo explico, porque a veces da lo mismo que en el ejemplo y otras veces no. La diferencia en "Frequency" es porque yo consider√© las compras con la misma fecha como solo 1 compra, en lugar de considerar cada item comprado como una compra. Ni idea por qu√© se obtienen resultados distintos en "Monetary", yo multipliqu√© el precio de los objetos por la cantidad que compr√≥ para obtener el dinero gastado, luego sum√© el dinero de una misma compra y luego promedi√© el dinero gastado en cada compra, me hace todo el sentido del mundo calcularlo de esta forma. En "Periodicity" nuevamente a veces se obtienen resultados iguales al ejemplo y a veces no.

### 1.3.1 Estandarizar Caracteristicas [0.5 puntos]

Construya una clase llamada ``MinMax()`` utilizando ``BaseEstimator`` y ``TransformerMixin`` para realizar una transformaci√≥n de cada una de las columnas de un DataFrame utilizando ``ColumnTransformer()`` m√°s tarde (tome como referencia el siguiente [enlace](https://sklearn-template.readthedocs.io/en/latest/user_guide.html#transformer)).


 Para esto considere que Min-Max escaler queda dada por la ecuaci√≥n:

$$MinMax = \dfrac{x-min(x)}{max(x) - min(x)}$$

Con esto buscamos que los valores que componen a las columnas se muevan en el rango de valores $[0, 1]$.

**Respuesta:**

```python
class MinMax(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        self.min = X.min()
        self.max = X.max()
        return self
    def transform(self, X):
        X2 = X.copy()
        return (X2 - self.min)/(self.max - self.min)
```

```python
ct = ColumnTransformer([('min-max', MinMax(), slice(1,6))])

pd.DataFrame(ct.fit_transform(output), columns=['Length', 'Recency', 'Frequency', 'Monetary', 'Periodicity'])
```

### 1.3.2 `T-SNA` Pipeline [1.0 puntos]

Para comenzar introduci√©ndose en el uso de pipeline, decide probar realizando un pipeline enfocado en la reducci√≥n de dimensionalidad y as√≠ hacer no decepcionar a Mr. Lepin con la clusterizaci√≥n del modelo. 

Configure un pipeline utilizando el algoritmo `T-SNE` sobre los datos **LRMFP**, donde, para la realizaci√≥n del pipeline considera los siguientes pasos:

1. Como primer paso obtenga las caracter√≠sticas **LRMFP** desde el DataFrame ``df_retail_II_cleaned.pickle`` utilizando la funci√≥n ``custom_features`` creada anteriormente, junto a ``FunctionTransformer()``. Considere esto como el primer paso de su pipeline.
2. En segundo lugar usando ``ColumnTransformer()`` aplique el MinxMax scaler creado por usted sobre todas las columnas generadas en el paso anterior. 
3. Finalmente, aplique un √∫ltimo paso donde obtiene las 2 componentes m√°s relevantes utilizando el algoritmo T-sne de sckit-learn.

Tras aplicar las transformaciones sobre el dataset **LRMFP**, gr√°fique las componentes obtenidas en la reducci√≥n de dimensionalidad.

**Respuesta:**

```python
pipe = Pipeline([('lrmfp', FunctionTransformer(custom_features)),
                 ('min-max', ColumnTransformer([('min-max', MinMax(), slice(1,6))])),
                 ('tsne', TSNE())
                ])
```

```python
result = pipe.fit_transform(df_retail)
result
```

### 1.3.3 Clustering

#### 1.3.3.1 M√©todo del Codo [1 puntos]

Utilizando la clase creada para escalamiento, aplique el m√©todo del codo para visualizar cual es el n√∫mero de clusters que mejor se ajustan a los datos. Realice esto utilizando el algoritmo K-means dentro de un pipeline para un $k \in [1,20]$, donde k representa el n√∫mero de clusters del k-means. Para la realizaci√≥n de esta secci√≥n y la pr√≥xima (1.3.3.2), considere los mismos pasos utilizados para el t-sne, pero permutando el algoritmo de reducci√≥n de dimensionalidad por k-means.

A trav√©s del grafico obtenido, comente y justifique que valor de k escoger√≠a para realizar el k-means.

**Respuesta:**

```python
pipe = Pipeline([('lrmfp', FunctionTransformer(custom_features)),
                 ('min-max', ColumnTransformer([('min-max', MinMax(), slice(1,6))]))])
data = pipe.fit_transform(df_retail)
l = []
n_clusters = list(range(1,21))
for k in n_clusters:
    cluster = KMeans(n_clusters=k).fit(data)
    l.append(cluster.inertia_)
```

```python
px.line(x=n_clusters, y=l)
```

Desde k=6 en adelante que la variaci√≥n en la inercia deja de ser tan notoria, por lo que se elige k=6 para clusterizar.

#### 1.3.3.2 Segmentaci√≥n de Clientes con K-Means üéÅ [1 punto]

En base a la elecci√≥n de k realizada en la secci√≥n anterior, utilice este valor escogido y entrene un modelo de K-means utilizando el mismo pipeline de scikit-learn utilizado anteriormente.

Una vez ajustado los datos, genere una tabla con los promedios (o medianas) para cada uno de los atributos, agrupando estos por el cl√∫ster que pertenecen. ¬øEs posible observar agrupaciones coherentes?, ¬øQu√© tipo de clientes posee el retail?, Justifique su respuesta y no decepcione a Mr. Lepin.


**Respuesta:**

```python
pipe = Pipeline([('lrmfp', FunctionTransformer(custom_features)),
                 ('min-max', ColumnTransformer([('min-max', MinMax(), slice(1,6))])),
                 ('k means', KMeans(n_clusters=6))])
cluster = pipe.fit_transform(df_retail)
```

```python
df = custom_features(df_retail).drop(columns=['Customer ID'])
df['Cluster'] = np.argmin(cluster, axis=1)
df
```

```python
round(df.groupby(by='Cluster').mean(),1)
```

Todas los clusters tienen cantidades muy similares de dinero gastado por compra, por lo que esto no es un buen indicador para distinguir entre clientes.

Se puede distinguir un grupo de clientes que es muy irregular para asistir a la tienda (cluster 5, periodicidad muy alta), un grupo de viejos clientes que va muy seguido (cluster 3, frecuencia muy alta), un grupo de clientes muy nuevos que ha ido poco y ha comenzado a asistir muy recientemente (cluster 4, frecuencia, periodicidad, largo y recencia muy bajos), un grupo de viejos clientes que no va muy seguido (cluster 0, frecuencia relativamente alta, periodicidad baja), un grupo de clientes que ya no asiste a esta tienda (cluster 2, recencia muy alta y largo corto) y un grupo de clientes que pareciera asistir regularmente pero cada largos periodos (cluster 1, largo, recencia y periodicidad relativamente altos, tercera frecuencia m√°s alta)
